```python
intent_keywords = {
    "nutrition": [
        "diet", "eat", "food", "fruits", "vegetables", "iron", "calcium", "sugar", "grains", "healthy", "pregnancy diet"
    ],
    "medicine": [
        "medicine", "paracetamol", "aspirin", "tablet", "dose", "safe", "treatment", "pain relief", "home remedy"
    ],
    "emergency": [
        "emergency", "hospital", "severe", "admit", "critical", "bleeding", "high fever", "chest pain"
    ],
    "symptoms": [
        "symptom", "signs", "cold", "cough", "fever", "rash", "headache", "covid", "dengue", "bp", "dehydration"
    ],
    "lifestyle": [
        "sleep", "mental", "exercise", "hydration", "screen time", "rest", "relieve", "good sleep", "stress", "routine"
    ]
}
intent_samples = {
    "nutrition": [
        "What to eat to increase iron?",
        "Healthy food during pregnancy",
        "Diet for diabetes control",
        "Foods that help with hemoglobin",
        "Calcium-rich vegetables"
    ],
    "medicine": [
        "Is paracetamol safe?",
        "Can I take aspirin during pregnancy?",
        "What medicines help cold symptoms?",
        "Painkillers for fever",
        "Home remedy for acidity"
    ],
    "emergency": [
        "When to go to hospital?",
        "What to do in emergency health cases?",
        "Is high fever dangerous?",
        "Should I admit for chest pain?",
        "Severe symptoms that need doctor"
    ],
    "symptoms": [
        "What are COVID-19 symptoms?",
        "Dengue signs and symptoms",
        "How does high BP feel?",
        "Symptoms of dehydration",
        "Cold and fever signs"
    ],
    "lifestyle": [
        "Tips for better sleep",
        "How to relieve period pain?",
        "How to manage stress?",
        "Mental health tips",
        "Reduce screen time for sleep"
    ]
}


def classify_keyword_based(user_input):
    user_input = user_input.lower()
    for intent, keywords in intent_keywords.items():
        if any(k in user_input for k in keywords):
            return intent
    return "unknown"

def classify_embedding_based(user_input):
    user_emb = model.encode(user_input, convert_to_tensor=True)
    best_score = -1
    best_intent = "unknown"
    for intent, samples in intent_samples.items():
        sample_embs = model.encode(samples, convert_to_tensor=True)
        score = util.cos_sim(user_emb, sample_embs).max().item()
        if score > best_score:
            best_score = score
            best_intent = intent
    return best_intent
```

```python
def chatbot_response(user_input, lang='en', threshold=0.6):
    user_emb = model.encode(user_input, convert_to_tensor=True)
    best_score = -1
    best_answer = "🤖 I don't know that exactly, but I'm learning!"

    for item in qa_data:
        q_emb = model.encode(item["question"], convert_to_tensor=True)
        score = util.cos_sim(user_emb, q_emb).item()
        if score > best_score and score > threshold:
            best_score = score
            best_answer = item.get(f"answer_{lang}", item["answer_en"])

    return best_answer
```

```python
# Demo chat loop (basic console-style interaction)
def run_chat():
    while True:
        user_input = input("Ask a health-related question (or type 'exit'): ")
        if user_input.lower() == 'exit':
            break
        print("\n🤖 Response:", chatbot_response(user_input, lang='en'))
        print("📌 Keyword Intent:", classify_keyword_based(user_input))
        print("🧠 Embedding Intent:", classify_embedding_based(user_input))
        print("—" * 50)

run_chat()
```
